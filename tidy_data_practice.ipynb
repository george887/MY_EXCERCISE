### Attendance Data

Load the attendance.csv file and calculate an attendnace percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

```python
# Accessing the csv file in my classifications_exersises untidy_data folder
df = pd.read_csv('/Users/georgearredondo/Desktop/codeup-data-science/classification_exercises/untidy_data/attendance.csv')
# Renaming the unnammed column to show student
df = df.rename(columns={'Unnamed: 0': 'student'})
df
```

```python
# Creating a dictionary to hold the attendance grades.
def letter_grade(letter):
    if letter == "P":
        return 1
    elif letter == "T":
        return .9
    elif letter == "H":
        return .5
    elif letter == "A":
        return 0
           
df = df.melt(id_vars = 'student', var_name = "date")
df['grade'] = df.value.apply(letter_grade)
df.groupby('student').grade.mean()
```

### 3. Cake Recipes

- Read the cake_recipes.csv data. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.
- Tidy the data as necessary.
- Which recipe, on average, is the best? recipe b
- Which oven temperature, on average, produces the best results? 275
- Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

```python
df3 = pd.read_csv('/Users/georgearredondo/Desktop/codeup-data-science/classification_exercises/untidy_data/cake_recipes.csv')
df3
# Need to seperate the recipe:positon also need to make the temperature a column
```

```python
# String split with expand = True to split into two seperat columns. Now rename 0 and 1
recipe = df3['recipe:position'].str.split(":", expand = True)
recipe
```

```python
# Renamed columns 0 and 1
recipe.columns = ["recipe", "position"]
recipe
```

```python
# Now need to add the columns to the original df.
# And now we can drop the recipe:position columm
df3 = pd.concat([df3,recipe], axis=1)
df3.head()
```

```python
df3 = df3.drop(columns = "recipe:position", axis=1)
df3
# Now we can make the temprature into a column
```

```python
# performed melt to seperate temperature. Now its nice and tidy
df3 = df3.melt(id_vars = ["recipe", "position"], var_name = 'temperature', value_name = "score")
df3.head()
```

```python
#Which recipe, on average, is the best? recipe b has an average score of 
round(df3.groupby('recipe').score.mean(),2)76.74
```

```python
# Which oven temperature, on average, produces the best results? 275 with 74.89
round(df3.groupby('temperature').score.mean(),2)
```

```python
# Which combination of recipe, rack position, and temperature gives the best result? 
# recipe b, bottom rack, 300 degrees
df3.sort_values(by = "score").tail(1)
```
