```python
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_people_who_died_in_traffic_collisions"
tables = pd.read_html(url)
```

```python
type(tables)
len(tables)
```

```python
df = tables[0]
df
```

```python
df.describe
df.dtypes
```

```python
# Extracting age out of Aged
age = [int(sub.split(' ')[0]) for sub in df.Aged]
age
```

```python
df['Age'] = age
df
```

```python
df.info()
```

```python
# Dropping null crash location
df = df.rename(columns={"Crash location": 'Crash_location'})
df = df.drop(columns={'Aged'})
```

```python
# looking at nulls
df.isna().sum()
```

```python
# Getting rid of the birth null value
df = df[df['Birth'].notna()]
df.info()
```

```python
df.isna().sum()
```

```python
df['Birth'] = df.Birth.astype('int64')
```

```python
df.head()
```

```python
df.info()
```

```python
df.Birth.astype('int64')
```

```python
type(df.Crash_location[0])
```

```python
df['Crash_location'] = df.Crash_location.astype(str)
```

```python
df.info()
```

```python
# Extracting the location to give me only the country
df['Country'] = df.apply(lambda row: row['Crash_location'].split(', ')[-1], axis=1)
```

```python
df.head()
```

```python
# Grouping by Country and keeping it a data frame
country = df.groupby('Country').count()[['Crash_location']]
```

```python
country
```

```python
country.columns = ['Events']
```

```python
country
```

```python
country['Total_Deaths'] = df.groupby('Country').sum()
```
